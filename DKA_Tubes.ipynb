```python
# prompt: make me a fuzzy with the code above

import numpy as np
import skfuzzy.control as ctrl
# Define fuzzy input variables
x_std_integrated = ctrl.Antecedent(np.arange(df[' Standard deviation of the integrated profile'].min(), df[' Standard deviation of the integrated profile'].max(), 0.1), 'std_integrated')
x_kurtosis_integrated = ctrl.Antecedent(np.arange(df[' Excess kurtosis of the integrated profile'].min(), df[' Excess kurtosis of the integrated profile'].max(), 0.1), 'kurtosis_integrated')
x_skewness_integrated = ctrl.Antecedent(np.arange(df[' Skewness of the integrated profile'].min(), df[' Skewness of the integrated profile'].max(), 0.1), 'skewness_integrated')
x_mean_dm_snr = ctrl.Antecedent(np.arange(df[' Mean of the DM-SNR curve'].min(), df[' Mean of the DM-SNR curve'].max(), 0.1), 'mean_dm_snr')
x_std_dm_snr = ctrl.Antecedent(np.arange(df[' Standard deviation of the DM-SNR curve'].min(), df[' Standard deviation of the DM-SNR curve'].max(), 0.1), 'std_dm_snr')
x_kurtosis_dm_snr = ctrl.Antecedent(np.arange(df[' Excess kurtosis of the DM-SNR curve'].min(), df[' Excess kurtosis of the DM-SNR curve'].max(), 0.1), 'kurtosis_dm_snr')
x_skewness_dm_snr = ctrl.Antecedent(np.arange(df[' Skewness of the DM-SNR curve'].min(), df[' Skewness of the DM-SNR curve'].max(), 0.1), 'skewness_dm_snr')
```

```python
# Define fuzzy output variable
y_target_class = ctrl.Consequent(np.arange(0, 1.1, 0.1), 'target_class')
```

```python
# Define membership functions (example, these should be tuned based on data analysis)
x_std_integrated.automf(3)
x_kurtosis_integrated.automf(3)
x_skewness_integrated.automf(3)
x_mean_dm_snr.automf(3)
x_std_dm_snr.automf(3)
x_kurtosis_dm_snr.automf(3)
x_skewness_dm_snr.automf(3)

y_target_class['pulsar'] = fuzz.trimf(y_target_class.universe, [0, 0, 1])
y_target_class['non_pulsar'] = fuzz.trimf(y_target_class.universe, [0, 1, 1])
```

```python
# Define fuzzy rules (example, these should be defined based on domain knowledge or learning)
# Rule 1: If skewness_integrated is high (good) AND skewness_dm_snr is high (good), then it is likely a pulsar
rule1 = fuzz.control.Rule(x_skewness_integrated['good'] & x_skewness_dm_snr['good'], y_target_class['pulsar'])

# Rule 2: If kurtosis_integrated is low (poor) AND kurtosis_dm_snr is low (poor), then it is likely a non_pulsar
rule2 = fuzz.control.Rule(x_kurtosis_integrated['poor'] & x_kurtosis_dm_snr['poor'], y_target_class['non_pulsar'])

# Rule 3: If std_integrated is high (good) AND std_dm_snr is high (good), then it is likely a non_pulsar (pulsars tend to have lower std dev)
rule3 = fuzz.control.Rule(x_std_integrated['good'] & x_std_dm_snr['good'], y_target_class['non_pulsar'])

# Rule 4: If mean_dm_snr is high (good), then it is likely a pulsar
rule4 = fuzz.control.Rule(x_mean_dm_snr['good'], y_target_class['pulsar'])

# Create the control system
pulsar_ctrl = fuzz.control.ControlSystem([rule1, rule2, rule3, rule4])

# Create the simulation object
pulsar_prediction = fuzz.control.ControlSystemSimulation(pulsar_ctrl)
```

Testing

```python
# To use the fuzzy system, you would set the input values and compute the output:
# For example, predicting for the first row of your dataframe:
pulsar_prediction.input['std_integrated'] = df.iloc[0][' Standard deviation of the integrated profile']
pulsar_prediction.input['kurtosis_integrated'] = df.iloc[0][' Excess kurtosis of the integrated profile']
pulsar_prediction.input['skewness_integrated'] = df.iloc[0][' Skewness of the integrated profile']
pulsar_prediction.input['mean_dm_snr'] = df.iloc[0][' Mean of the DM-SNR curve']
pulsar_prediction.input['std_dm_snr'] = df.iloc[0][' Standard deviation of the DM-SNR curve']
pulsar_prediction.input['kurtosis_dm_snr'] = df.iloc[0][' Excess kurtosis of the DM-SNR curve']
pulsar_prediction.input['skewness_dm_snr'] = df.iloc[0][' Skewness of the DM-SNR curve']

# Compute the result
pulsar_prediction.compute()

print(pulsar_prediction.output['target_class'])

# You would typically iterate through your dataset or apply this to new data
# to get predictions.
```

The result seen above is a result of the prediction of fuzzy. If it shown value more than 0.5, the star is classified as pulsar star.
